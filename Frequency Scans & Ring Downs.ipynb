# Frequency Scans

```python
import numpy as np
import matplotlib.pyplot as plt

freq = np.linspace(32768-10, 32768+10, 2000)
scan_amp = .5/np.sqrt(.25+(freq-32768)**2)

plt.figure(1,figsize = (15,10))
plt.plot(np.linspace(32768,32768,10), np.linspace(0.0099,1,10), 
         color = 'black', label = '$f_0 = 32768$ Hz',linewidth = 2)
plt.plot(freq,scan_amp,label = 'Simulated Data',linewidth = 3)
plt.rcParams['font.size'] = '19'
plt.xticks(np.arange(min(freq), max(freq)+1, 2.0))
plt.xlabel('Frequency [Hz]')
plt.ylabel('Fractional Amplitude')
plt.title('Simulated Frequency Scan')
plt.legend()
```

# Ring Downs

```python
A = 1
gamma = .7
nat_freq = 32768
delta = np.pi/4
time = np.linspace(0,10,50000)
rd_amp = A*np.exp(-gamma*time/2)*np.cos(nat_freq*time +delta)

plt.figure(1,figsize = (15,10))
plt.rcParams['font.size'] = '19'
plt.plot(time,rd_amp)
plt.plot([-1,10],[0,0], color ='black')
plt.xlim(-0.1,10)
plt.title('Simulated Ring Down')
plt.xlabel('Time [s]')
plt.ylabel('Fractional Amplitude')
```

```python
plt.figure(1,figsize = (15,10))
plt.rcParams['font.size'] = '19'
plt.plot(time,rd_amp)
plt.plot([-1,10],[0,0], color ='black')
plt.xlim(0,.2)
plt.title('Simulated Ring Down')
plt.xlabel('Time [s]')
plt.ylabel('Fractional Amplitude')
```
